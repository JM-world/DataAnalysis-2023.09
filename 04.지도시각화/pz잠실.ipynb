- 영등포구 소재 지하철역

```python
import folium
import pandas as pd
```

```python
df = pd.read_csv('data/잠실2.csv')
df
```

```python
map = folium.Map(location=[df['위도'].mean(), df['경도'].mean()], zoom_start=14)    # Center position
for i in df.index:
    folium.Marker(
        location=[df['위도'][i],df['경도'][i]],
        tooltip=df['이름'][i],
        popup=folium.Popup(df['주소'][i], max_width=200)
    ).add_to(map)
title_html = '<h3 align= "center" style="font-size:20px"> 잠실 명소</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```

```python
df['color'] = ['red', 'blue', 'green', 'violet', 'orange']
df
```

```python
map = folium.Map(location=[df['위도'].mean(), df['경도'].mean()], zoom_start=14)    # Center position
for i in df.index:
    folium.Marker(
        location=[df['위도'][i], df['경도'][i]],
        icon = folium.CustomIcon(f'data/{i}.jpg'), icon_size=(50,50)
    ).add_to(map)

title_html = '<h3 align= "center" style="font-size:20px"> 잠실명소</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
