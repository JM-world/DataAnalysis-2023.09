## Pandas의 시각화 기능

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

- Line graph

```python
np.random.seed(0)
df = pd.DataFrame(np.random.randn(100, 3), columns=list('ABC'),
                    index=pd.date_range('2023-09-01', periods=100))      # 3개의 열에 100개의 데이터가 들어온다.
df.tail()
```

```python
df = df.cumsum()
df.tail()
```

```python
df.plot(figsize=(5, 3))         # default가 line graph
plt.title('Pandas의 plot method 사용 예')
plt.xlabel('일자'), plt.ylabel('데이터')
plt.show()
```

- Bar chart

```python
iris = sns.load_dataset('iris')
iris.sepal_length[:30].plot(kind='bar', rot=45, figsize=(6, 3)) # rot = x축 글자 기울기
plt.title('Iris sepal length 데이터 30개');
```

```python
iris.sepal_length[:30].plot(kind='barh')
plt.title('Iris sepal length 데이터 30개');
```

```python
iris.sepal_length[:30].sort_values(ascending=False).plot(kind='barh') # sort_values: 값 순서대로
plt.title('Iris sepal length 데이터 30개');
```

```python
df2 = iris.groupby('species').mean()
df2
```

```python
df2.plot.bar(figsize=(6, 3), rot=0)
plt.title('Iris 품종별 피처의 평균');
```

```python
df.columns.name='Feature'               # 범례에 제목 생성
df2.plot.bar(figsize=(6, 3), rot=0)
plt.ylim(0, 10)                         # 범례가 그림과 겹치지 않게
plt.title('Iris 품종별 피처의 평균');
```

- Pie chart

```python
titanic = sns.load_dataset('titanic')

s = titanic['class'].value_counts()
s
```

```python
s.sort_index().plot.pie(autopct = '%.1f%%')
plt.title('선실별 승객 수');
```

- Histogram

```python
titanic.age.plot.hist(figsize=(4, 3))
plt.title('타이타닉호의 탑승객의 연령');
```

- Box plot

```python
iris[iris.species == 'setosa'].plot.box(figsize=(4,3))
plt.title('Setosa 품종의 각 피처별 Box plot');
plt.show()
```

```python
iris.boxplot(by='species')
plt.suptitle('Iris 각 피쳐의 품종별 박스 플롯');
```
