### Matplotlib 기본 Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False      # x, y축 레이블에 - 깨지는 현상 방지
```

### 1. 라인 플롯

```python
plt.figure(figsize=(4, 2))
plt.plot([0,1,2,3], [0,1,4,9])
plt.title('라인 플롯'), plt.grid()#; 을 붙여주고 plt.show()를 제외해도 된다.
plt.show()
```

- 스타일 지정

```python
plt.figure(figsize=(4, 2))
plt.plot([0,1,4,9], 'rs--')     # r(red)컬러, s(square)마커, linestyle
plt.title("'rs--' 스타일의 플롯"), plt.grid()
plt.show()
```

```python
plt.figure(figsize=(4, 2))
plt.plot([0,1,4,9], 'mo-.')
plt.plot([9,4,1,0], 'g^:')
plt.title("'mo-.','g^:' 스타일의 플롯"), plt.grid()
plt.show()
```

```python
plt.figure(figsize=(4, 2))
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r") #lw=라인 굵기
plt.title("스타일 적용 예")
plt.grid()
plt.show()
```

- 그림 범위

```python
plt.figure(figsize=(4, 2))
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50), plt.ylim(-10, 30), plt.grid()
plt.show()
```

- 곡선 그리기

```python
plt.figure(figsize=(4, 3))
X = np.linspace(0, 2 * np.pi, 200)  # 라인을 몇 등분 할 것인지
Ysin, Ycos = np.sin(X), np.cos(X)
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Sin, Csine 그래프'), plt.grid()
plt.show()
```

- Tick 설정

```python
plt.figure(figsize=(4, 3))
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('Tick 설정'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
plt.yticks([-1, -0.707, 0, 0.707, 1])
plt.show()
```

```python
plt.figure(figsize=(4, 3))
plt.plot(X, Ysin), plt.plot(X, Ycos)
plt.title('LaTeX, 문자열로 tick label 정의'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])    # LaTeX
plt.yticks([-1, -0.707, 0, 0.707, 1],
           [r'$-1$', r'$-\sqrt{2}/2$', r'$0$', r'$\sqrt{2}/2$', r'$1$'])
plt.show()
```

- 범례(legend) 표시

```python
plt.figure(figsize=(4, 3))
plt.plot(X, Ysin, label='사인'), plt.plot(X, Ycos, label='코사인')
plt.title('범례 표시'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])    # LaTeX
plt.yticks([-1, -0.707, 0, 0.707, 1],
           [r'$-1$', r'$-\sqrt{2}/2$', r'$0$', r'$\sqrt{2}/2$', r'$1$'])
plt.legend(loc='upper center')      # 파라메터를 생략하면 적절한 위치에 표시해줌
plt.show()
```

- X축, Y축 Label

```python
plt.figure(figsize=(4, 3))
plt.plot(X, Ysin, label='사인'), plt.plot(X, Ycos, label='코사인')
plt.title('X축, Y축 label 표시'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])    # LaTeX
plt.yticks([-1, -0.707, 0, 0.707, 1],
           [r'$-1$', r'$-\sqrt{2}/2$', r'$0$', r'$\sqrt{2}/2$', r'$1$'])
plt.xlabel('시간'), plt.ylabel('진폭')
plt.legend() 
plt.show()
```

- 제목 크기 지정

```python
plt.figure(figsize=(4, 3))
plt.plot(X, Ysin, label='사인'), plt.plot(X, Ycos, label='코사인')
plt.title('그림 크기, 제목 크기 지정', fontsize= 15)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])    # LaTeX
plt.yticks([-1, -0.707, 0, 0.707, 1],
           [r'$-1$', r'$-\sqrt{2}/2$', r'$0$', r'$\sqrt{2}/2$', r'$1$'])
plt.xlabel('시간'), plt.ylabel('진폭')
plt.legend(), plt.grid()
plt.show()
```

- 그림 저장

```python
plt.figure(figsize=(4, 3))
plt.plot(X, Ysin, label='사인'), plt.plot(X, Ycos, label='코사인')
plt.title('그림 저장', fontsize= 15)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])    # LaTeX
plt.yticks([-1, -0.707, 0, 0.707, 1],
           [r'$-1$', r'$-\sqrt{2}/2$', r'$0$', r'$\sqrt{2}/2$', r'$1$'])
plt.xlabel('시간'), plt.ylabel('진폭')
plt.legend(), plt.grid()
plt.savefig('data/삼각함수.png')
```
