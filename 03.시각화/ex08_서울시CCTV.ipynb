## 서울시 CCTV - 인구 시각화

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('../02.Pandas/data/서울시_구별_CCTV_인구_현황.csv',index_col='구분')
df.head()
```

#### 1. 자치구별 CCTV 설치 댓수

```python
df.CCTV댓수.plot(figsize=(6, 5), kind='barh')
plt.title('자치구별 CCTV 설치 댓수')
plt.show()
```

```python
# CCTV 댓수로 정렬하고, plt.barh()로 그리기
sc = df['CCTV댓수'].sort_values()
plt.barh(sc.index, sc.values)
plt.title('자치구별 CCTV 설치 댓수 정렬완료');
```

```python
# Seaborn으로 그리기
sns.barplot(data=df, y = df.index, x = 'CCTV댓수');
```

```python
scr = df['CCTV댓수'].sort_values(ascending=False)
sns.barplot(y = scr.index, x = scr.values)
plt.title('서울시 구별 CCTV 설치댓수 정렬완료');
```

##### 2. 구별 인당 CCTV 댓수 그리기 - plt.barh()

```python
ic = (df['CCTV댓수'] / df['인구수']).sort_values()
plt.barh(ic.index , ic.values);
```

##### 3. 구별 외국인당 CCTV 댓수 - Seaborn

```python
wi = (df['외국인'] / df['CCTV댓수']).sort_values(ascending=False)
sns.barplot(x = wi.values, y = wi.index)
plt.title('구별 외국인당 CCTV 댓수');
```

##### 4. 인구수와 CCTV 댓수 분석

- 산점도

```python
plt.figure(figsize=(6,6))
plt.scatter(df['인구수'], df['CCTV댓수'])
plt.title('인구수와 CCTV 댓수의 관계')
plt.grid(), plt.xlabel('인구수'), plt.ylabel('CCTV 댓수');
```

- 회귀선

```python
fp1 = np.polyfit(df['인구수'], df['CCTV댓수'], 1)
fp1             # [기울기, 절편]
```

```python
fx = np.array([100000, 700000])
fy = fx * fp1[0] + fp1[1]
fy
```

```python
f1 = np.poly1d(fp1)
fy = f1(fx)
fy
```

```python
plt.figure(figsize=(4, 4))
plt.scatter(df['인구수'], df['CCTV댓수'])
plt.plot(fx, fy, 'g--', lw=3)
plt.title('인구수와 CCTV 댓수의 관계', fontsize= 15)
plt.grid(), plt.xlabel('인구수'), plt.ylabel('CCTV 댓수');
```

- 오차의 절댓값

```python
df['오차'] = abs(df['CCTV댓수'] - f1(df['인구수'])).round(1)
df.head(3)
```

```python
# 오차의 내림차순으로 정렬
df_sort = df.sort_values(by='오차', ascending=False).head(10)
df_sort
```

- 최종 그림

```python
plt.figure(figsize=(12, 9))
plt.scatter(df['인구수'], df['CCTV댓수'], c=df['오차'])
plt.plot(fx, fy, 'g--', lw=3)

for i in range(10):
    if i == 8:
        continue
    plt.text(df_sort['인구수'][i]+8000, df_sort['CCTV댓수'][i]-50, df_sort.index[i], fontsize=12)
plt.text(df_sort['인구수'][8]-5000, df_sort['CCTV댓수'][8]-200, df_sort.index[8], fontsize=13)
plt.title('인구수와 CCTV 댓수의 관계', fontsize= 20)
plt.grid(), plt.xlabel('인구수'), plt.ylabel('CCTV 댓수')
plt.colorbar();
```
