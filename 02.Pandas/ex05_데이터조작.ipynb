- 정렬: sort_values(), sort_index()

```python
np.random.seed(2023)
df = pd.DataFrame(np.random.randint(40,101, size=(4, 5)), index=list('abcd'), columns=list('VWXYZ'))
df
```

```python
# W 컬럼 오름차순 기준
df.sort_values(by='W')
```

```python
# Y 컬럼 내림차순 기준, Y 컬럼에 동일한 값이 있으면 Z컬럼의 내림차순
df2 = df.sort_values(by=['Y','Z'], ascending=False)
df2
```

```python
df2.sort_index()
```

- apply 변환

```python
np.random.seed(2023)
df = pd.DataFrame(np.random.randint(40,101, size=(4, 5)), index=list('abcd'), columns=list('VWXYZ'))
df
```

```python
# 각 열에서 최댓값 - 최솟값 한 결과
df.apply(lambda x: x.max() - x.min())   # x는 각 column을 의미 
```

```python
# V 열의 제곱수
df.V.apply(lambda x: x*x)   # x는 V열의 각 엘리먼트
```

```python
# 각 행에서 최댓값 - 최솟값 한 결과 - axis=1을 주면 됨
df.apply(lambda x: x.max() - x.min(), axis = 1)
```

```python
# 타이타닉호의 승객 중 나이 20살을 기준으로 성인(adult)과 미성년자(child)를 구별
titanic['adult/child'] = titanic.age.apply(lambda x: 'adult' if x >=20 else 'child')
titanic.tail()
```

```python
data = {
    'text': ['03.입출력.ipynb','04.고급인덱싱.ipynb','05.데이터조작.ipynb'],
    'price': ['24,000','36,000','42,000']
}
df1 = pd.DataFrame(data)
df1
```

```python
# 구둣점 제거 열 생성
df1['구둣점 제거'] = df1.text.str.replace('.', ',')
df1
```

```python
# 한글 텍스트 열 생성
df1['한글 텍스트'] = df1.text.str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣]','', regex=True)
df1
```

```python
# price 문자열을 정수로 변환하여 가격 열 생성
df1['가격'] = df1.price.apply(lambda x: int(x.replace(',','')))
df1
```
